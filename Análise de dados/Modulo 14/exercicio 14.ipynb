## 1\. Preço do diamante por tipo de corte

Grafico de Barra

```
import seaborn as sns
import matplotlib.pyplot as plt


diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```
diamonds = diamonds[["cut", "price"]].groupby("cut").agg("sum").reset_index()
diamonds.head()
```

```
grafico = sns.barplot(data=diamonds, x="cut", y="price", ci=None, palette="pastel")
grafico.set(title='Distribuição de Preços por Tipo de Corte', xlabel='Tipo de Corte', ylabel='Preço ($)');
```

```
grafico = sns.barplot(data=diamonds, y="cut", x="price", ci=None, palette="pastel")
grafico.set(title='Distribuição de Preços por Tipo de Corte', xlabel='Tipo de Corte', ylabel='Preço ($)');
```

**Insight do gráfico 1:**

A análise dos preços dos diamantes em relação ao tipo de corte revela insights valiosos sobre as preferências e valorizações do mercado de joias.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

december_data = flights_data[flights_data['month'] == 'Dec']
december_data.head()
```

```
import matplotlib.pyplot as plt

with sns.axes_style('whitegrid'):

  grafico = sns.FacetGrid(december_data,  palette="pastel")
  grafico.map(sns.lineplot, "year", "passengers")
  grafico.map(plt.fill_between, 'year', 'passengers', alpha=0.3)
  grafico.set(title='Passageiros por ano em dezembro', xlabel='Ano', ylabel='Passageiros');
  grafico.fig.set_size_inches(w=15/2.54, h=7.5/2.54)
```

**Insight do gráfico 2**: ...

Ao analisar os insights sobre o número de passageiros em voos durante o mês de dezembro ao longo dos anos, observa-se uma tendência ascendente significativa. Essa análise revela um aumento constante no volume de passageiros de um ano para o outro.

## 3\. Numero de passageiros por mês entre 1949 e 1959

grafico de linha

```
import seaborn as sns

flights = sns.load_dataset("flights")

data = flights.pivot(index="year", columns="month", values="passengers")
data.head()
```

```
grafico = sns.lineplot(data=flights, x="year", y="passengers", hue="month", style="month")
grafico.set(title='Passageiros por mês entre 1949 e 1959', xlabel='Mês', ylabel='Passageiros');
grafico.get_legend().set_title("Ano");
```

**Insight do gráfico 3**: ...

 Entre os anos de 1949 e 1959, a demanda por transporte aéreo aumentou consistentemente, refletindo a crescente popularidade das viagens aéreas. Esse fenômeno pode ser atribuído a diversos fatores, como o desenvolvimento tecnológico na aviação, a expansão das rotas aéreas e a melhoria na acessibilidade aos serviços de transporte aéreo.
